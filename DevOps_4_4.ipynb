## DevOps

**DevOps** — это сочетание двух слов: Development (Разработка) и Operations (Операции). **DevOps** — это практика, которая позволяет одной команде управлять всем жизненным циклом разработки приложения, включая разработку, тестирование, развертывание и мониторинг.

**Основная цель** DevOps — уменьшить продолжительность жизненного цикла разработки системы, одновременно часто поставляя новые функции, исправления и обновления в тесной синхронизации с бизнес-целями.

DevOps — это подход к разработке программного обеспечения, с помощью которого можно разрабатывать качественное ПО быстро и с большей надежностью. Он включает различные этапы, такие как непрерывная разработка, непрерывная интеграция, непрерывное тестирование, непрерывное развертывание и непрерывный мониторинг.

### История DevOps

До появления DevOps существовало два подхода к разработке программного обеспечения: каскадная модель (Waterfall) и Agile.

#### **Каскадная модель (Waterfall Model)**

**Каскадная модель** — это линейная и довольно простая модель разработки программного обеспечения, которая следует подходу «сверху вниз».

Она состоит из нескольких этапов, начиная со сбора и анализа требований. На этом этапе команда получает требования от клиента для разработки приложения. После этого проводится анализ этих требований.

Следующий этап — этап проектирования (Design), где создаётся схема программного обеспечения. Здесь разрабатывается представление о том, как программа будет выглядеть и функционировать.

После того как проектирование завершено, наступает этап реализации (Implementation), на котором начинается кодирование приложения. Команда разработчиков совместно работает над различными компонентами приложения.

После завершения разработки приложение тестируется на этапе верификации (Verification). На этом этапе проводятся различные виды тестирования, такие как модульное тестирование, интеграционное тестирование, тестирование производительности и др.

После того как все тесты пройдены, приложение развёртывается на рабочих серверах.

И наконец, наступает этап поддержки (Maintenance). На этом этапе осуществляется мониторинг производительности приложения, и все проблемы, связанные с его производительностью, решаются.

**Преимущества каскадной модели:**

* Простота в понимании и использовании
* Облегчает тестирование и анализ
* Экономит значительное количество времени и денег
* Подходит для небольших проектов, если все требования четко определены
* Позволяет организовать процесс по отделам и обеспечивает управленческий контроль

**Недостатки каскадной модели:**

* Риски и неопределенность
* Недостаточная видимость текущего прогресса
* Непригодна, если требования постоянно меняются
* Сложно внести изменения в продукт, когда он уже находится на стадии тестирования
* Конечный продукт доступен только по завершении цикла
* Не подходит для крупных и сложных проектов

#### **Методология Agile**

**Agile** — это итеративный подход к разработке программного обеспечения, при котором проект разбивается на отдельные итерации или спринты. Каждая итерация включает этапы, аналогичные этапам каскадной модели, такие как сбор требований, проектирование, разработка, тестирование и поддержка. Длительность каждой итерации обычно составляет от 2 до 8 недель.

image.png

**Процесс Agile**

В Agile-компании приложение выпускается с некоторыми высокоприоритетными функциями уже в первой итерации.

После выпуска конечные пользователи или клиенты предоставляют обратную связь о производительности приложения.

Затем в приложение вносятся необходимые изменения и добавляются новые функции, после чего оно снова выпускается — это уже вторая итерация.

Этот процесс повторяется до тех пор, пока не будет достигнуто необходимое качество программного обеспечения.

**Преимущества модели Agile:**

* Адаптивно реагирует на изменения требований
* Исправление ошибок на ранних этапах делает процесс более экономичным
* Улучшает качество продукта и минимизирует количество ошибок
* Обеспечивает прямую коммуникацию между участниками проекта
* Хорошо подходит для крупных и долгосрочных проектов
* Требует минимальных ресурсов и легко управляется

**Недостатки модели Agile:**

* Сильно зависит от четкости требований клиента
* Достаточно сложно предсказать время и усилия для крупных проектов
* Непригодна для сложных проектов
* Недостаточная эффективность документации
* Увеличенные риски поддерживаемости

### Этапы и инструменты DevOps

Как упоминалось ранее, этапы непрерывной разработки, непрерывной интеграции, непрерывного тестирования, непрерывного развертывания и непрерывного мониторинга составляют жизненный цикл DevOps. Рассмотрим каждый из этих этапов жизненного цикла DevOps по порядку.

**Этап 1:** Непрерывная разработка (Continuous Development)

Используемые инструменты: Git, SVN, Mercurial, CVS.

Процесс:

image.png

На этом этапе выполняются планирование и кодирование программного обеспечения. На стадии планирования определяется видение проекта, и разработчики начинают писать код для приложения.

Для планирования инструменты DevOps не требуются, но существует множество инструментов для поддержания кода.

Код может быть написан на любом языке, но для его поддержания используются инструменты контроля версий. Этот процесс называется управлением исходным кодом (Source Code Management).

После завершения разработки кода начинается этап непрерывной интеграции (Continuous Integration).

**Этап 2:** Непрерывная интеграция (Continuous Integration)

Инструменты: Jenkins, TeamCity, Travis.

Процесс:

Этот этап является ключевым в жизненном цикле DevOps. На практике разработчики должны регулярно вносить изменения в исходный код, обычно ежедневно или еженедельно.

Каждое изменение (коммит) затем проходит сборку, что позволяет выявлять возможные проблемы на ранней стадии. Сборка кода включает не только компиляцию, но и обзор кода, модульное тестирование, интеграционное тестирование и упаковку.

image.png

Код, поддерживающий новые функции, постоянно интегрируется с существующим кодом. Поскольку программное обеспечение непрерывно развивается, необходимо интегрировать обновленный код без перерывов и слаженно с другими системами, чтобы изменения были отражены для конечных пользователей.

На этом этапе используются инструменты для сборки и упаковки кода в исполнимый файл, чтобы затем передать его на следующие этапы жизненного цикла DevOps.

**Этап 3:** Непрерывное тестирование (Continuous Testing)

Инструменты: Jenkins, Selenium, TestNG, JUnit.

Процесс:

image.png

На этом этапе программное обеспечение тестируется непрерывно на наличие ошибок с помощью инструментов автоматизированного тестирования. Эти инструменты позволяют тестировщикам проверять несколько кодовых баз параллельно, чтобы убедиться в отсутствии функциональных ошибок. Для симуляции тестовой среды на этом этапе можно использовать контейнеры Docker.

Selenium используется для автоматизированного тестирования, а отчеты генерируются с помощью TestNG. Весь процесс тестирования можно автоматизировать с помощью инструмента непрерывной интеграции Jenkins.

Предположим, вы написали код для тестирования приложения с помощью Selenium на языке Java. Этот код можно собрать с использованием инструментов, таких как Ant или Maven. После сборки кода, его можно протестировать на соответствие требованиям пользовательского тестирования (User Acceptance Testing, UAT). Этот процесс также можно автоматизировать с помощью Jenkins.

**Этап 4:** Непрерывное развертывание (Continuous Deployment)

Используемые инструменты:

* Управление конфигурациями — Chef, Puppet, Ansible.

* Контейнеризация — Docker, Vagrant.


Процесс:

На этом этапе код развертывается на производственных серверах. Важно также убедиться, что код корректно развернут на всех серверах. Прежде чем продолжить, давайте разберемся с некоторыми инструментами управления конфигурациями и контейнеризации. Эти инструменты помогают в достижении непрерывного развертывания (CD).

image.png

**Управление конфигурациями** — это процесс установления и поддержания согласованности функциональных требований и производительности приложения. Проще говоря, это процесс развертывания на серверах, планирования обновлений на всех серверах и, что наиболее важно, поддержания一 конфигураций на всех серверах.

Инструменты контейнеризации также играют ключевую роль на этапе развертывания. Эти инструменты помогают обеспечивать согласованность между средами разработки, тестирования, предварительного развертывания и производства. Кроме того, они помогают быстро масштабировать экземпляры, увеличивая или уменьшая их количество по мере необходимости.

**Этап 5:** Непрерывный мониторинг (Continuous Monitoring)

Используемые инструменты: Splunk, ELK Stack, Nagios, New Relic.

Процесс:

Это очень важный этап в жизненном цикле DevOps, на котором вы постоянно мониторите производительность вашего приложения. На этом этапе записывается важная информация о использовании программного обеспечения. Затем эта информация обрабатывается для проверки корректности работы приложения. В рамках этого этапа устраняются системные ошибки, такие как низкий уровень памяти, недоступность сервера и другие.

image.png

Эта практика включает участие команды операций, которая будет отслеживать активность пользователей на наличие ошибок или ненормального поведения системы. Инструменты для непрерывного мониторинга помогают следить за производительностью приложения и серверами, а также позволяют проактивно проверять состояние системы.

### Кто такой инженер DevOps?

Инженер DevOps — это специалист, который понимает жизненный цикл разработки программного обеспечения и имеет глубокие знания различных инструментов автоматизации для создания цифровых пайплайнов (CI/CD пайплайнов).

Инженер DevOps работает с разработчиками и IT-специалистами для контроля над релизами кода. Это могут быть разработчики, которые заинтересовались развертыванием и операциями с сетью, или системные администраторы, увлекающиеся написанием скриптов и кодированием, которые переходят на сторону разработки, где могут улучшать планирование тестирования и развертывания.

image.png

### Над какими факторами разработчики могут работать?

Что ж, разработчики могут работать в основном над тремя важными факторами:

* Программное обеспечение, которое поставляется.

* Эффективность и производительность среды, в которую поставляется программное обеспечение.

* Повышение эффективности, функциональности и скорости разработки программного обеспечения при меньших затратах.

Таким образом, когда вы поставляете программное обеспечение, вы не просто передаёте его в производство, а выполняете весь жизненный цикл поставки программного обеспечения.

image.png

Как видно на схеме, разработчики создают программное обеспечение и хранят его в системе управления конфигурацией программного обеспечения или в системе контроля версий. После этого среды контроля качества проверяют качество, проводится системное интеграционное тестирование и пользовательское приемочное тестирование. Наконец, когда программное обеспечение проходит все эти этапы, оно попадает в производство, где фактически работает и с ним взаимодействуют клиенты.

Теперь среды, в которых работают разработчики, могут совпадать со средами, в которых работают клиенты, но конфигурации могут отличаться. Поэтому для соответствия этим конфигурациям требуется автоматическое развёртывание.

**Чем DevOps отличается от традиционных IT?**

Традиционные ИТ-системы состоят из тысяч строк кода и создаются разными командами с разными стандартами, в то время как DevOps-системы создаются одной командой, которая хорошо разбирается в продукте. Традиционные ИТ-системы сложны для понимания, а DevOps-системы легко понять.

**10 Лучших инструментов DevOps**

1. Дженкинс
Цель: Непрерывная интеграция и непрерывная поставка (CI /CD)
Особенности: обширная экосистема плагинов, автоматизация процессов сборки и развертывания, а также масштабируемость.
2. Docker
Цель: Контейнеризация
Особенности: упрощает развертывание приложений, обеспечивает согласованность в разных средах и поддерживает архитектуру микросервисов.
3. Kubernetes
Назначение: Управление контейнерами
Особенности: автоматизирует развертывание, масштабирование и управление контейнерными приложениями.
4. Terraform
Назначение: Инфраструктура как код (IaC)
Функции: управляет инфраструктурой в нескольких облачных средах, обеспечивает воспроизводимость и поддерживает контроль версий.
5. Ansible
Назначение: Управление конфигурацией и автоматизация
Особенности: простой синтаксис (YAML), бессерверная архитектура и масштабируемость.
6. Прометей
Назначение: Мониторинг и оповещение
Функции: сбор и хранение показателей, мощный язык запросов (PromQL) и интеграция с Grafana для визуализации.
7. GitLab
Назначение: управление исходным кодом и CI / CD
Особенности: интегрированная платформа DevOps, поддерживающая управление версиями кода, конвейеры CI/CD и инструменты управления проектами.
8. ELK Stack (Elasticsearch, Logstash, Kibana)
Назначение: ведение журнала и аналитика
Функции: поиск и анализ данных в реальном времени, объединение журналов и мощные возможности визуализации.
9. Azure DevOps
Назначение: комплексное решение DevOps
Особенности: поддерживает CI/CD, управление проектами и интегрируется с различными инструментами разработки и мониторинга.
10. Nagios
Назначение: Мониторинг инфраструктуры
Функции: отслеживает сетевые службы, ресурсы сервера и предупреждает о возможных проблемах.


**Преимущества DevOps**
- Более быстрая доставка: DevOps позволяет организациям выпускать новые продукты и обновления быстрее и чаще, что может обеспечить конкурентное преимущество.
- Улучшенная совместная работа: DevOps способствует совместной работе команд разработчиков и операторов, что приводит к улучшению коммуникации, повышению эффективности и снижению трения.
- Улучшенное качество: DevOps делает упор на автоматизированное тестирование и непрерывную интеграцию, что помогает выявлять ошибки на ранних этапах разработки и повышать общее качество программного обеспечения.
- Повышенная автоматизация: DevOps позволяет организациям автоматизировать многие процессы, выполняемые вручную, высвобождая время для более стратегической работы и снижая риск человеческих ошибок.
- Лучшая масштабируемость: DevOps позволяет организациям быстро и эффективно масштабировать свою инфраструктуру в соответствии с меняющимися требованиями, улучшая способность реагировать на потребности бизнеса.
- Повышение удовлетворенности клиентов: DevOps помогает организациям быстрее выпускать новые функции и обновления, что может привести к повышению удовлетворенности и лояльности клиентов.
- Повышенная безопасность: DevOps способствует внедрению передовых методов обеспечения безопасности, таких как непрерывное тестирование и мониторинг, которые помогают снизить риск нарушений безопасности и повысить общую безопасность систем организации.
- Более эффективное использование ресурсов: DevOps позволяет организациям оптимизировать использование ресурсов, включая оборудование, программное обеспечение и персонал, что может привести к экономии средств и повышению эффективности.


**Проблемы при внедрении DevOps**
- Высокие первоначальные затраты: внедрение DevOps может быть сложным и дорогостоящим процессом, требующим значительных вложений в технологии, инфраструктуру и персонал.
- Нехватка специалистов: поиск квалифицированных специалистов по DevOps может оказаться непростой задачей, и организациям, возможно, придётся инвестировать в программы обучения и развития, чтобы сформировать необходимые навыки в своих командах.
- Сопротивление изменениям: некоторые сотрудники могут сопротивляться культурным и организационным изменениям, необходимым для успешного внедрения DevOps, что может привести к сопротивлению, нежеланию сотрудничать и снижению эффективности.
- Отсутствие стандартизации: DevOps — всё ещё относительно новая область, и в ней не хватает стандартизации в отношении методологий, инструментов и процессов. Из-за этого организациям может быть сложно определить наилучший подход для своих конкретных нужд.
- Повышенная сложность: DevOps может усложнить процесс разработки программного обеспечения, требуя от организаций управления большим количеством компонентов и интеграции нескольких систем и инструментов.
- Зависимость от технологий: DevOps в значительной степени зависит от технологий, и организациям может потребоваться инвестировать в различные инструменты и платформы для поддержки процесса DevOps.
- Необходимость постоянного совершенствования: DevOps требует постоянного совершенствования и адаптации по мере появления новых технологий и передовых практик. Организации должны быть готовы постоянно адаптировать и развивать свои методы DevOps, чтобы оставаться конкурентоспособными.


#### СЕРИАЛИЗАЦИЯ И ДЕСЕРИАЛИЗАЦИЯ


Код, который был написан для обучения модели и оценки её качества, крайне редко используется для инференса (от англ inference — вывод). Так называется непрерывная работа алгоритма машинного обучения в конечном приложении. По этой причине при внедрении моделей в продакшн их принято сохранять в готовом виде, то есть уже обученными и готовыми решать реальные задачи.

image.png

Как и почти всё в языке программирования Python, обученная модель является **объектом**. Этот объект не простой, поскольку модель содержит сложную иерархию классов — в каждом классе есть набор полей, ссылающихся на объекты других классов, и так далее.

Например, объект класса `RandomForestClassifier` из библиотеки sklearn содержит множество полей, часть из которых устанавливается во время инициализации модели (максимальная глубина, количество деревьев в ансамбле, критерий информативности и т. д.), а часть определяется во время обучения модели (последовательность предикатов внутри каждого дерева в ансамбле, значимость признаков). Наша задача — «законсервировать» этот объект (модель), сохранив значение всех полей, которые мы задали при инициализации объекта и получили по итогам обучения. То есть мы должны сохранить модель, включая её внешние и внутренние параметры.

Чтобы гарантировать сохранение всей структуры данных и получить её при загрузке обратно, используется **сериализация**.

> **Сериализация** — это процесс преобразования структуры данных или объекта в формат, который можно сохранить и передать. Результатом этого процесса является последовательность байтов, представляющая объект в "плоском" виде, удобном для хранения и передачи. Обратный процесс — **десериализация** — это восстановление объекта из байтовой последовательности в его исходное состояние.

Цель сериализации заключается в том, чтобы зафиксировать состояние объекта для дальнейшего использования.



image.png

**Основные принципы сериализации**

1. *Сохранение структуры:* Сериализация должна сохранить внутреннюю структуру и значения объекта. Например, для сериализации списка в Python каждый элемент списка и его порядок должны быть сохранены.

2. *Поддержка типов данных:* Процесс сериализации должен учитывать тип данных каждого элемента структуры. Это важно для правильного восстановления объекта в исходном виде.

3. *Платформенная независимость:* Многие форматы сериализации предназначены для работы на разных платформах и языках программирования, чтобы можно было передавать данные между системами.

**Зачем нужна десериализация?**

1. *Восстановление состояния объектов:* После сохранения данных в файл или передачи их через сеть десериализация позволяет восстановить объекты и их состояние для продолжения работы с ними.

2. *Работа с переданными данными:* В распределённых системах данные передаются в сериализованном виде. Десериализация на стороне получателя позволяет интерпретировать и обработать их.

3. *Обработка данных из внешних источников:* Данные, сохранённые в формате JSON, XML или Protocol Buffers, можно десериализовать для обработки в системе или приложении, независимо от их источника.

#### Зачем это нужно?

То, в каком виде данные записаны на диск, зачастую может сильно отличаться от того, в каком виде они существуют в памяти программы. И там, и там это будут байты информации, но в файле будет представлена просто их последовательность, а в памяти программы это может быть какой-то объект или даже несколько объектов со структурой.

Приведём простой пример с форматом CSV. Пусть у нас есть файл с одной строкой, где несколько слов записаны через запятую. На диске это будет записано просто как последовательность байтов.

Однако когда мы считываем эту последовательность внутри программы, мы хотим работать с ней как со **списком строк**, то есть нам нужно не только считать эти данные, но и применить к ним некоторое преобразование:

```python
line = 'word1, word2, word3'

line.split(",")
```

Можно заметить, что после этого преобразования появилось много объектов (каждое слово списка) и некоторая структура (то, как слова расположены в этом списке). Процесс, который мы только что выполнили, называется **десериализацией**.

Также возможен обратный процесс, когда мы хотим сохранить в виде последовательности байтов какие-то данные из памяти программы. Например, нам нужно снова получить начальную строку:

```python
",".join(['word1', 'word2', 'word3'])
```

Этот процесс как раз и называется **сериализацией**.

Мы разобрали простейший пример, но на практике всё гораздо сложнее. Сериализовывать можно не только в текст (как с CSV, JSON и подобными форматами), но и в **бинарный** формат, который человек не сможет прочитать.

Бывают форматы, которые могут описывать более сложные структуры (тот же JSON). Также можно добавить сжатие итогового набора битов.

Заметим, что программа должна потратить некоторый ресурс CPU, чтобы преобразовать объект в набор байтов и наоборот.

#### ИНСТРУМЕНТЫ СЕРИАЛИЗАЦИИ: PICKLE

Посмотрим, с помощью каких средств происходит сериализация объектов в Python.

В стандартную библиотеку Python входит модуль pickle, который служит для сериализации почти всех объектов произвольного типа.

Мы помним, что объекты находятся в оперативной памяти и направляются в байтовые потоки ввода-вывода. В байтовые потоки может быть направлен любой файлоподобный объект.

В ходе десериализации исходный объект воссоздаётся в оперативной памяти с теми же самыми значениями, но с новой идентичностью — новым адресом в памяти.



image.png

Так как законсервирован может быть абсолютно любой объект, в нём могут быть «спрятаны» различные вредоносные программы или данные. Поэтому будьте внимательны и не проводите десериализацию бинарных файлов, в происхождении которых вы не уверены.

Для иллюстрации работы модуля pickle последовательно пройдёмся по всем шагам и разберёмся с тем, как сериализовать уже обученные алгоритмы машинного обучения.

#### **Шаг №1**

Обучим модель линейной регрессии на встроенном датасете о диабете — Diabetes dataset.

В данном датасете представлены десять исходных признаков: возраст, пол, индекс массы тела, среднее артериальное давление и шесть измерений сыворотки крови были получены для каждого из 442 пациентов с сахарным диабетом. Интерес представляет количественный показатель прогресса заболевания, замеренный через год после исходного измерения. Тип задачи — регрессия.

*Примечание*. Для простоты и наглядности мы опустим процесс предобработки данных, разведывательный анализ, разделение выборки на обучающую и тестовую, валидацию модели и подбор гиперпараметров, так как для наших целей это неважно.

В качестве модели, прогнозирующей целевую переменную, возьмём простейшую линейную регрессию, и обучим её на исходных данных:

```python
from sklearn.linear_model import LinearRegression

from sklearn.datasets import load_diabetes

# Загружаем датасет о диабете
X, y = load_diabetes(return_X_y=True)
# Инициализируем модель линейной регрессии
regressor = LinearRegression()
# Обучаем модель
regressor.fit(X,y)

## LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

В результате выполнения кода получился объект класса `LinearRegression`, на который ссылается переменная `regressor`. При этом атрибуты объекта (веса модели линейной регрессии) были сформированы во время обучения. То есть объект `regressor` теперь является обученной моделью.

#### **Шаг №2**

Далее, когда мы получили обученную модель, нам необходимо сериализовать её, превратив объект Python в поток байтов. Для этого импортируем модуль pickle и воспользуемся функцией `dumps()`, в которую нужно передать объект Python.

```python
import pickle

# Производим сериализацию обученной модели
model = pickle.dumps(regressor)

print(type(model))
print(type(regressor))
## bytes
## sklearn.linear_model._base.LinearRegression
```

#### **Шаг №3**

Давайте попробуем восстановить (десериализовать) объект Python. Для этого в модуле pickle есть функция `loads()`, в которую нужно передать сериализованный объект (поток байтов).

```python
# Производим десериализацию
regressor_from_bytes = pickle.loads(model)
regressor_from_bytes
## LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

В результате десериализации мы смогли восстановить исходный объект (модель).

#### **Шаг №4**

Сохраним сериализованный объект прямо в файл. Для этого в pickle есть функция `dump()` (без s на конце). В неё необходимо передать имя файла или ссылку на открытый файл. Файл назовём myfile, его расширение — .pkl (формат данных pickle):

```python
# Производим сериализацию и записываем результат в файл формата pkl
with open('myfile.pkl', 'wb') as output:
    pickle.dump(regressor, output)
```

#### **Шаг №5**

Посмотрим на код, который восстанавливает (десериализует) обученную модель из файла myfile.pkl. Для этого в pickle есть функция `load()` (без s на конце). В неё необходимо передать имя файла или ссылку на открытый файл.

```python
# Производим десериализацию и извлекаем модель из файла формата pkl
with open('myfile.pkl', 'rb') as pkl_file:
    regressor_from_file = pickle.load(pkl_file)

regressor_from_file
## LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

#### **Шаг №6**

Убедимся, что методы и результаты предсказаний обученной модели и модели, загруженной из файла, совпадают:

```python
# Проверяем, что все элементы массивов предсказаний совпадают между собой
print(all(regressor.predict(X) == regressor_from_bytes.predict(X)))
## True
print(all(regressor.predict(X) == regressor_from_file.predict(X)))
## True
```

Как видите, исходная и восстановленная из байтов и файла модели дают одинаковые предсказания. Это значит, что теперь мы можем импортировать наши обученные модели в любое Python-приложение и пользоваться ими, минуя этап обучения и все этапы, предшествующие ему.

##### Вопросы:

#### СОХРАНЕНИЕ ПАЙПЛАЙНА

Ранее мы посмотрели простейший пример сериализации готовой модели.

У вас мог возникнуть вопрос: что делать, если перед подачей данных в модель их необходимо предобработать, например произвести стандартизацию, исключить неинформативные признаки? Неужели придётся прописывать все эти шаги в коде инференса модели? А что если вопросами инференса занимаются совершенно другие специалисты, которые вообще ничего не знают о машинном обучении и не умеют производить предобработку данных?

Конечно, мы должны передать результаты в таком виде, чтобы ими можно было воспользоваться без лишних манипуляций.

Мы уже упоминали, что pickle работает с любыми объектами Python. Поэтому для сохранения может быть доступна не просто обученная модель, но и целый **пайплайн**, включающий предобработку данных.

> **Пайплайн** — это автоматизированный поэтапный процесс выполнения манипуляций с данными, включающий в себя сбор, обработку, генерацию и отбор признаков, обучение модели с последующей её настройкой и проверкой качества.

Например, мы хотим сериализовать пайплайн, который включает в себя min-max-нормализацию и отбор пяти наиболее важных факторов на основе корреляции Пирсона. Полученные в результате данные отправляются на вход модели линейной регрессии.

```python
import pickle

from sklearn.linear_model import LinearRegression

from sklearn.datasets import load_diabetes

from sklearn.feature_selection import SelectKBest, f_regression

from sklearn.preprocessing import MinMaxScaler

from sklearn.pipeline import Pipeline

# Загружаем датасет о диабете
X, y = load_diabetes(return_X_y=True)

# Создаём пайплайн, который включает нормализацию, отбор признаков и обучение модели
pipe = Pipeline([
  ('Scaling', MinMaxScaler()),
  ('FeatureSelection', SelectKBest(f_regression, k=5)),
  ('Linear', LinearRegression())
  ])

# Обучаем пайплайн
pipe.fit(X, y)
```

Пайплайн обучен. Давайте сохраним его в файл с помощью pickle:

```python
# Сериализуем pipeline и записываем результат в файл
with open('my_pipeline.pkl', 'wb') as output:
    pickle.dump(pipe, output)
```

Если сериализация завершилась успешно, то при инференсе модели мы сможем восстановить её из файла:

```python
# Десериализуем pipeline из файла
with open('my_pipeline.pkl', 'rb') as pkl_file:
    loaded_pipe = pickle.load(pkl_file)
```

Проверим, что результаты исходного и десериализованного пайплайнов и идентичны:

```python
# Сравниваем предсказания исходного и восстановленного пайплайнов
print(all(pipe.predict(X) == loaded_pipe.predict(X)))

## True
```

*Примечание*. Если мы хотим сохранять сериализованные пайплайны в виде потока байтов, нужно использовать функции `dumps()` и `loads()`, а не `dump()` и `load()`.

Однако в процессе предобработки могут возникнуть шаги, которые нельзя реализовать стандартными методами sklearn. Например, для решения многих задач часто использовался feature engineering, чтобы повысить качество работы моделей. Как встроить этот шаг в исходный пайплайн?

Для этого в sklearn можно организовать так называемые **кастомные трансформеры**.

> **Кастомные трансформеры** — это пользовательские классы, которые позволяют создавать свои собственные преобразования данных для использования в пайплайнах обработки данных и машинного обучения. Они полезны, когда стандартных трансформеров, доступных в библиотеке, например, в scikit-learn, недостаточно для выполнения нужных преобразований.



#### Как работают кастомные трансформеры?

В scikit-learn для создания кастомного трансформера нужно унаследовать класс от BaseEstimator и TransformerMixin.

> **BaseEstimator** — это класс, от которого наследуются все основные модели и трансформеры scikit-learn. Этот класс предоставляет некоторые базовые функции, упрощающие работу с гиперпараметрами, настройками и стандартизацией.

> **TransformerMixin** — это миксин, который предоставляет метод fit_transform(). Этот метод сразу выполняет и fit(), и transform() за один вызов, что полезно для трансформеров, которые сначала обучаются на данных, а затем сразу применяют преобразования. Благодаря TransformerMixin кастомный трансформер может использовать fit_transform(), что делает его совместимым с пайплайнами scikit-learn.



Посмотрим на шаблон кастомного трансформера:

```python
from sklearn.base import TransformerMixin, BaseEstimator


class MyTransformer(TransformerMixin, BaseEstimator):
    '''Шаблон кастомного трансформера'''

    def __init__(self):
        '''
        Здесь прописывается инициализация параметров, не зависящих от данных.
        '''
        pass

    def fit(self, X, y=None):
        '''
        Здесь прописывается «обучение» трансформера.
        Вычисляются необходимые для работы трансформера параметры (если они нужны).
        '''

        return self

    def transform(self, X):
        '''
        Здесь прописываются действия с данными.
        '''
        return X
```

У трансформера должно быть **три обязательных метода**:

* __init__() — метод, который вызывается при создании объекта данного класса. Он предназначен для инициализации исходных параметров.
Например, у трансформера для создания полиномиальных признаков PolynomialFeatures из sklearn в методе `__init__()` параметр `degree` задаёт степень полинома.

* fit() — метод, который вызывается для «обучения» трансформера. Он должен возвращать ссылку на сам объект (`self`).
Например, в трансформере StandardScaler в методе `fit()` прописано вычисление среднего значения и стандартного отклонения в каждом столбце таблицы, переданной в качестве параметра метода `fit()`.

* transform() — метод, который трансформирует приходящие на вход данные. Он должен возвращать преобразованный массив данных.
Например, при вызове метода `transform()` у StandardScaler из sklearn внутри происходит преобразование — вычитание из каждого столбца среднего и деление результата на стандартное отклонение. Причём среднее и стандартное отклонение вычисляются заранее в методе `fit()`.

*Примечание*. У некоторых трансформеров из sklearn, например у того же MinMaxScaler, есть ещё и метод `fit_transform()`, который является комбинацией методов `fit()` и `transform()`.

Наш трансформер пока что ничего не делает. Предположим, мы хотим генерировать в данных новый признак, который является простым произведением первых трёх столбцов таблицы. Давайте пропишем в методе `transform()` эти действия.

Для работы такого трансформера нужны только исходные данные без дополнительных параметров, поэтому методы `__init__()` и `fit()` остаются без изменений.

```python
import numpy as np

class MyTransformer(TransformerMixin, BaseEstimator):
    '''Шаблон кастомного трансформера'''


    def __init__(self):
        '''Здесь прописывается инициализация параметров, не зависящих от данных.'''
        pass


    def fit(self, X, y=None):
        '''
        Здесь прописывается «обучение» трансформера.
        Вычисляются необходимые для работы трансформера параметры (если они нужны).
        '''
        return self


    def transform(self, X):
        '''Здесь прописываются действия с данными.'''
        # Создаём новый столбец как произведение первых трёх
        new_column = X[:, 0] * X[:, 1] * X[:, 2]
        # Для добавления столбца в массив нужно изменить его размер на (n_rows, 1)
        new_column = new_column.reshape(X.shape[0], 1)
        # Добавляем столбец в матрицу измерений
        X = np.append(X, new_column, axis=1)
        return X
```

Посмотрим, как работает наш кастомный трансформер. Создадим объект трансформера, вызовем метод `transform` и посмотрим на результирующий размер таблицы.

```python
# Инициализируем объект класса MyTransformer (вызывается метод __init__)
custom_transformer = MyTransformer()
# Чисто формально вызываем метод fit, но у нас он ничего не делает
custom_transformer.fit(X)
# Трансформируем исходные данные (вызывается метод transform)
X_transformed = custom_transformer.transform(X)
print('Shape before transform: {}'.format(X.shape))
print('Shape after transform: {}'.format(X_transformed.shape))

## Shape before transform: (442, 10)
## Shape after transform: (442, 11)
```

Видно, что в результате трансформации в исходную матрицу наблюдений добавился новый столбец.

Теперь давайте встроим этот трансформер в сам пайплайн — для этого достаточно добавить новый шаг в пайплайн.

```python
# Создаём пайплайн, который включает Feature Engineering, нормализацию, отбор признаков и обучение модели
pipe = Pipeline([
  ('FeatureEngineering', MyTransformer()),
  ('Scaling', MinMaxScaler()),
  ('FeatureSelection', SelectKBest(f_regression, k=5)),
  ('Linear', LinearRegression())
  ])

# Обучаем пайплайн
pipe.fit(X, y)
```
